```python
import pandas as pd
import numpy as np

data=pd.read_csv('pokemon.csv',header=0)
to_remove=['abilities','base_egg_steps','base_happiness','capture_rate','classfication','experience_growth',
          'height_m','japanese_name','percentage_male','pokedex_number','type1','type2','weight_kg','generation']
to_keep=[x for x in data.columns if x not in to_remove]
data=data[to_keep]
tier_listing={}
tier_listing['Aegislash']=1
tier_listing['Arceus']=1
tier_listing['Blaziken']=1
tier_listing['Darkrai']=1
tier_listing['Deoxys']=1
tier_listing['Dialga']=1
tier_listing['Genesect']=1
tier_listing['Giratina']=1
tier_listing['Groudon']=1
tier_listing['Ho-oh']=1
tier_listing['Kyogre']=1
tier_listing['Kyurem']=1
tier_listing['Landorus']=1
tier_listing['Lugia']=1
tier_listing['Lunala']=1
tier_listing['Marshadow']=1
tier_listing['Mewtwo']=1
tier_listing['Naganadel']=1
tier_listing['Palkia']=1
tier_listing['Pheromosa']=1
tier_listing['Rayquaza']=1
tier_listing['Reshiram']=1
tier_listing['Solgaleo']=1
tier_listing['Xerneas']=1
tier_listing['Yveltal']=1
tier_listing['Zekrom']=1
tier_listing['Blacephalon']=2
tier_listing['Celesteela']=2
tier_listing['Chansey']=2
tier_listing['Clefable']=2
tier_listing['Excadrill']=2
tier_listing['Ferrothorn']=2
tier_listing['Garchomp']=2
tier_listing['Gastrodon']=2
tier_listing['Gliscor']=2
tier_listing['Greninja']=2
tier_listing['Hawlucha']=2
tier_listing['Heatran']=2
tier_listing['Jirachi']=2
tier_listing['Kartana']=2
tier_listing['Keldeo']=2
tier_listing['Magearna']=2
tier_listing['Magnezone']=2
tier_listing['Mew']=2
tier_listing['Mimikyu']=2
tier_listing['Pelipper']=2
tier_listing['Serperior']=2
tier_listing['Skarmory']=2
tier_listing['Tangrowth']=2
tier_listing['Tapu Bulu']=2
tier_listing['Tapu Fini']=2
tier_listing['Tapu Koko']=2
tier_listing['Tapu Lele']=2
tier_listing['Toxapex']=2
tier_listing['Tyranitar']=2
tier_listing['Victini']=2
tier_listing['Volcarona']=2
tier_listing['Zapdos']=2
tier_listing['Zygarde']=2
num_pokemon=data.shape[0]
tier_vals=np.zeros(num_pokemon)
for i in range(num_pokemon):
    tier_vals[i]=tier_listing.get(data['name'][i],3)
data_final=data.assign(tier=tier_vals)
to_keep=[x for x in data_final.columns if x != 'name']
data_final=data_final[to_keep]
```

```python
X = data_final.loc[:, data_final.columns != 'tier']
y = data_final.loc[:, data_final.columns == 'tier']

from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(X_train, y_train)

# error rate on train set
pred_y_train=clf.predict(X_train)
y_train_wPred=y_train.assign(pred_tier=pred_y_train)
num_train=y_train_wPred.shape[0]
listOne=y_train_wPred['tier'].tolist()
listTwo=y_train_wPred['pred_tier'].tolist()
error_list=[listOne[i]==listTwo[i] for i in range(num_train)]
error_rate=error_list.count(0)/num_train
print(y_train_wPred)
print(error_rate)

# error rate on test set
pred_y_test=clf.predict(X_test)
y_test_wPred=y_test.assign(pred_tier=pred_y_test)
num_test=y_test_wPred.shape[0]
listOne=y_test_wPred['tier'].tolist()
listTwo=y_test_wPred['pred_tier'].tolist()
error_list=[listOne[i]==listTwo[i] for i in range(num_test)]
error_rate=error_list.count(0)/num_test
print(y_test_wPred)
print(error_rate)
```
